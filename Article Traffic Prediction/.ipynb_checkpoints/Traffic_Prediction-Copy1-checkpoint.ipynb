```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# define plt settings
sns.set_theme()
plt.rcParams["font.size"] = 20
plt.rcParams["axes.labelsize"] = 20
plt.rcParams["xtick.labelsize"] = 20
plt.rcParams["ytick.labelsize"] = 20
plt.rcParams["legend.fontsize"] = 20
plt.rcParams["legend.markerscale"] = 1.5
plt.rcParams["figure.figsize"] = (20, 10)
plt.rcParams["legend.title_fontsize"] = 20
```

```python
dataset=pd.read_csv("Datasets/Dataset.csv")
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
```

```python
dataset['Date']= le.fit_transform(dataset['Date'])
print(le.classes_)
dataset = dataset.drop("Day", axis=1)
```

```python
dataset.head(10)
```

```python
dataset.info()
```

```python
# show pairplot of the features
plt.figure(figsize=(10,10))
sns.pairplot(data=dataset, hue="Traffic", palette='pastel', height=5)
```

```python
plt.figure(figsize=(12,10))
cor = dataset.corr()
sns.heatmap(cor, annot=True,
            cmap=plt.cm.Greens, vmin=-1, vmax=1)
plt.show()
```

```python
X = dataset.drop('Traffic', axis=1)
t = dataset['Traffic']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, t_train, t_test = train_test_split(X, t, test_size = 0.20, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, t_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
if(y_pred.all()<2.5):
    y_pred=np.round(y_pred-0.5)
else:
    y_pred=np.round(y_pred+0.5)
```

```python
df1=(y_pred-t_test)/t_test
df1=round(df1.mean()*100,2)
print("Error = ",df1,"%") 
a=100-df1
print("Accuracy= ",a,"%")
```

```python
from sklearn import svm
clf_svm = svm.SVC()
clf_svm.fit(X_train, t_train)
```

```python
y_pred = clf_svm.predict(X_test)
y_pred
```

```python
if(y_pred.all()<2.5):
    y_pred=np.round(y_pred-0.5)
    
else:
    y_pred=np.round(y_pred+0.5)

y_pred
```
